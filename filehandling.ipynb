```python
class Detail():
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def collect(self):
        f= open('File.docx', 'w')
        f.write(f'Name: {self.name}\n')
        f.write(f'Age: {self.age}\n')
        f.close()   
    def display(self):
        f= open('File.docx', 'r')
        print(f.read())
        f.close()   

x=Detail('Himanshu', 20)
x.collect()
x.display()
```
