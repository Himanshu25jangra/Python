```python
class Atm(object):
    def __init__(self) -> None:
        print("Database conntected")
    
    def withdrawl(self):
        print("withdraw")
    def deposit(self):
        print("deposit")
    def minist(self):
        print("withdraw")
    def Bal(self):
        print("deposit")
    def menu(self):
        x= int(input("Press 1. withdrawl  2. deposit 3. minist 4. Bal"))
        match x:
            case 1:
                self.withdrawl()
            case 2:
                self.deposit()
            case 3:
                self.minist()
            case 4:
                self.Bal()
            case _:
                print("Invalid option")

atm = Atm()
atm.menu()
```

```python
class Father:
    def show(self):
        print("Parent Class")

class Son(Father):
    def disp(self):
        print("Child class")
        
x= Father()
x.show()

y=Son()
y.show()
y.disp()
```

```python
from abc import ABC,abstractmethod
class Father(ABC):
    @abstractmethod
    def disp(self):
        pass
    def show(self):
        print("Parent Class")

class Son(Father):
    def disp(self):
        print("Child class")
        

y=Son()
y.show()
y.disp()
```

```python
class Employee:
    def __init__(self, name, age, designation):
        self.name = name
        self.age = age
        self.designation = designation.lower()
        self.salary = self.get_salary()

    def get_salary(self):
        salary_structure = {
            'programmer': 25000,
            'manager': 30000,
            'tester': 20000
        }
        return salary_structure.get(self.designation, None)

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}, Designation: {self.designation.capitalize()}, Salary: â‚¹{self.salary}")


class EmployeeManager:
    def __init__(self):
        self.employee_list = []

    def add_employee(self):
        name = input("Enter name: ")
        age = input("Enter age: ")
        designation = input("Enter designation (programmer/manager/tester): ")

        emp = Employee(name, age, designation)

        if emp.salary is None:
            print("Error: Invalid designation. Please enter programmer, manager, or tester.")
        else:
            self.employee_list.append(emp)

    def display_all_employees(self):
        print("\nEmployee Details:")
        for emp in self.employee_list:
            emp.display()

    def run(self):
        while True:
            self.add_employee()
            cont = input("Do you want to add another employee? (yes/no): ").lower()
            if cont != 'yes':
                break
        self.display_all_employees()


# Run the program
if __name__ == "__main__":
    manager = EmployeeManager()
    manager.run()
```
